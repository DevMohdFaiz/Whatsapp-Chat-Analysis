```python
from chat_analyzer import WhatsAppChatAnalyzer
# Or use: from src.chat_analyzer import WhatsAppChatAnalyzer

# Initialize analyzer with the actual file path
analyzer = WhatsAppChatAnalyzer('../chats/WhatsApp Chat - The Shakirs.zip', chat_type='auto')

# Load and analyze
print("Loading and parsing...")
raw_df = analyzer.load_and_parse()
print(f"Raw DataFrame shape: {raw_df.shape}")

print("\nPreprocessing...")
processed_df = analyzer.preprocess()
print(f"Processed DataFrame shape: {processed_df.shape}")

print("\nEnriching with features...")
enriched_df = analyzer.enrich()
print(f"Enriched DataFrame shape: {enriched_df.shape}")
print(f"Columns: {list(enriched_df.columns)}")

print("\nRunning full analysis...")
results = analyzer.analyze()
print(f"\nAnalysis Results:")
print(f"  Chat type: {results['chat_type']}")
print(f"  Total messages: {results['total_messages']}")
print(f"  Date range: {results['date_range']}")
print(f"  Statistics: {results['statistics']}")

# Generate visualizations
print("\nGenerating visualization...")
fig = analyzer.visualize('message_frequency', by='sender')
fig.show()

# Get AI insights (requires GROQ_API_KEY in environment)
# insights = analyzer.get_ai_insights('summary', period='all')
# print(f"\nAI Insights: {insights}")
```
